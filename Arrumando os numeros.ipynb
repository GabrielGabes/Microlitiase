```python
#PACOTES
import pandas as pd #manipulação de dados
import numpy as np #matematica

import warnings #Warning inuteis
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

import unidecode #arrumar strings

import re #regex

import os
```

```python
df = pd.read_excel('ECOS - MICROLITIASE GERAL.xlsx')
df
```

```python
df.columns
```

```python
df['Idade'] = np.nan
for i in range(len(df)):
    if not pd.isnull(df['DADOS DO TASY'][i]):
        texto = df['DADOS DO TASY'][i]
        texto = texto[:texto.find('Médico assistente')]
        
        email = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b', texto)
        telefones = re.findall(r'\(\d{2}\) \d{9}', texto)
        idade = re.findall(r'(\d{1,3}a)', texto)    
        
        print(i)
        print(telefones)
        print(email)
        print(idade)
        print('\n')
        
        if len(idade) >= 1:
            df['Idade'][i] = idade[0]
        
        if len(email) == 1:
            df['Email'][i] = email[0]
            
        if len(telefones) > 1:
            df['Tel_resi'][i] = telefones[0]
            df['Tel_par'][i] = telefones[1]
            
        if len(telefones) == 1:
            df['Tel_resi'][i] = telefones[0]
        
```

```python
df.to_excel('ECOS - MICROLITIASE (feito6).xlsx')
```
